```python
import pandas as pd
import matplotlib.pyplot as plt



```

```python
df=pd.read_csv("ecommerce_sales.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
```

```python
df.groupby("Category")["Sales"].sum().plot(kind='bar', title="Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.month

df.groupby("Month")["Sales"].sum().plot(marker='o', title="Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False).plot(kind='bar', title="Top Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind='pie', autopct='%1.1f%%', title="Sales by Region")
plt.ylabel("")
plt.show()
```
